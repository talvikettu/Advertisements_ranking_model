```python
import pandas as pd
import numpy as np
import gc
import lightgbm as lgb
from sklearn.model_selection import GroupKFold
```

```python
use_cols = [
    "query_id", "item_id",
    "query_text", "item_title", "item_description",
    "query_cat", "query_mcat", "query_loc",
    "item_cat_id", "item_mcat_id", "item_loc",
    "price", "item_query_click_conv", "item_contact"
]

train = pd.read_parquet("train-dset.parquet", columns=use_cols)
print(train.shape)
```

```python
def build_light_features(df):
    df = df.copy()

    df["price"] = df["price"].fillna(df["price"].median())

    mean_price = df.groupby("query_id")["price"].transform("mean")

    X = pd.DataFrame({
        "cat_match": (df["query_cat"] == df["item_cat_id"]).astype("int8"),
        "mcat_match": (df["query_mcat"] == df["item_mcat_id"]).astype("int8"),
        "loc_match": (df["query_loc"] == df["item_loc"]).astype("int8"),

        "log_price": np.log1p(df["price"]).astype("float32"),
        "price_ratio_mean": (df["price"] / (mean_price + 1e-6)).astype("float32"),
        "price_diff_mean": (df["price"] - mean_price).astype("float32"),

        "item_query_click_conv": df["item_query_click_conv"].fillna(0).astype("float32"),

        "len_query": df["query_text"].str.len().astype("float32"),
        "len_title": df["item_title"].str.len().astype("float32"),
        "len_desc": df["item_description"].str.len().astype("float32"),
    })

    return X
```

```python
X = build_light_features(train)
y = train["item_contact"].astype("int8")
groups = train.groupby("query_id").size().values

print(X.shape)
```

```python
del train
gc.collect()
```

```python
def calc_dcg_at_k(v, k=10):
    w = 0.97 ** np.arange(len(v))
    return (v * w)[:k].sum()

def calc_ndcg_at_k(v, k=10):
    return calc_dcg_at_k(v, k) / calc_dcg_at_k(np.sort(v)[::-1], k)
```

```python
gkf = GroupKFold(n_splits=3)

ndcg_scores = []
feature_importances = []

for fold, (train_idx, val_idx) in enumerate(gkf.split(X, y, groups=groups)):
    print(f"\nFold {fold+1}")

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    group_train = groups[train_idx]
    group_val = groups[val_idx]

    model = lgb.LGBMRanker(
        objective="lambdarank",
        metric="ndcg",
        n_estimators=300,
        learning_rate=0.05,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

    model.fit(
        X_train, y_train,
        group=group_train,
        eval_set=[(X_val, y_val)],
        eval_group=[group_val],
        eval_at=[10],
        verbose=50
    )

    y_pred = model.predict(X_val)

    val_df = pd.DataFrame({
        "query_id": np.repeat(np.unique(groups[val_idx]), group_val),
        "y_true": y_val.values,
        "y_pred": y_pred
    })

    ndcg_list = []
    for qid, group_df in val_df.groupby("query_id"):
        ndcg_list.append(calc_ndcg_at_k(group_df.sort_values("y_pred", ascending=False)["y_true"].values))

    mean_ndcg = np.mean(ndcg_list)
    print("NDCG@10:", mean_ndcg)
    ndcg_scores.append(mean_ndcg)

    feature_importances.append(model.feature_importances_)
```

```python
print("Mean NDCG@10:", np.mean(ndcg_scores))
```

```python
feat_imp = pd.DataFrame(feature_importances, columns=X.columns).mean().sort_values(ascending=False)

feat_imp_df = feat_imp.reset_index()
feat_imp_df.columns = ["feature", "importance"]

feat_imp_df
```
